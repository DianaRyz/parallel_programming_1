```python
import matplotlib.pyplot as plt

sizes = []
times = []
with open("D:/parallel programming/result/stats.txt", 'r') as file:
    for line in file:
        size, time = map(int, line.strip().split('\t'))
        sizes.append(size)
        times.append(time)


plt.plot(sizes, times, marker='o', linestyle='-')
plt.title('Среднее время')
plt.xlabel('Размер матриц')
plt.ylabel('Время (мс)')
plt.grid(True)
plt.show()
```
